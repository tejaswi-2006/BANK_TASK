```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")


# Load dataset
df = pd.read_csv('/content/drive/MyDrive/DataAnalysis/data_YesBank_StockPrices.csv')

print("Data Loaded Successfully")
print(df.head())


# Convert Date column to datetime (fixes .dt error)
df['Date'] = pd.to_datetime(
    df['Date'],
    format='%b-%y',
    errors='coerce'
)

# Remove invalid dates (if any)
df = df.dropna(subset=['Date'])

# Sort by Date
df = df.sort_values('Date')


# Feature Engineering
df['Range'] = df['High'] - df['Low']          # Volatility
df['Return'] = df['Close'].pct_change() * 100 # Monthly Return
df['Year'] = df['Date'].dt.year               # Extract Year


# Reset index
df = df.reset_index(drop=True)


print("\nPrepared Data Preview:")
print(df.head())

print("\nData Types:")
print(df.dtypes)
```

```python
#Close Price Histogram
sns.histplot(df['Close'], kde=True)
plt.title("Distribution of Close Price")
plt.show()
```
